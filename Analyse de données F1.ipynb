```python
from matplotlib import pyplot as plt
import fastf1 as ff1
import fastf1.plotting
import pandas as pd
import os
ff1.plotting.setup_mpl()
import numpy as np
```

```python
cache_dir = os.path.expanduser('~/.fastf1/cache')
os.makedirs(cache_dir, exist_ok=True)

# Activer le cache de FastF1
ff1.Cache.enable_cache(cache_dir)
```

```python
fp1 = ff1.get_session(2023,'Miami', 'FP1')
fp2 = ff1.get_session(2023,'Miami', 'FP2')
#fp3 = ff1.get_session(2023,'Australia', 'FP3')
```

```python
fp1_laps = fp1.load_laps(with_telemetry = True)
fp2_laps = fp2.load_laps(with_telemetry = True)
#fp3_laps = fp3.load_laps(with_telemetry = True)
```

```python
fp1_laps.dropna(subset = ["LapTime"])
```

```python
fp1_laps = fp1_laps.dropna(subset = ["LapTime"])

drivers = np.array(fp1_laps.Driver.tolist())
drivers = np.unique(drivers)
tyre_type = ["SOFT", "MEDIUM", "HARD"]

drivers_time = {key:None for key in drivers}

fastest_lap = fp1_laps.pick_fastest().LapTime.total_seconds()

for driver in drivers:
    data = fp2_laps.pick_driver(driver).pick_tyre("HARD")
    if len(data) != 0:
        data["LapTime"] = data["LapTime"].apply(lambda x: x.total_seconds())
        data = data.drop(data[data['LapTime'] >= fastest_lap+10].index)
        drivers_time[driver] = data['LapTime'].mean()
    
display(drivers_time)
```

```python
drivers_time = {k: v for k, v in sorted(drivers_time.items(), key=lambda item: item[1], reverse=False)}

# Obtenir les clés et les valeurs triées
categories = list(drivers_time.keys())
values = list(drivers_time.values())
fig = plt.figure(figsize = (12,8))
# Tracer le graphique avec des barres décroissantes
plt.bar(categories, values, color = 'r')

# Afficher le graphique
plt.show()
```

```python
import fastf1 as ff1
from fastf1 import plotting
from fastf1 import utils
from matplotlib import pyplot as plt

plotting.setup_mpl()

session = ff1.get_session(2021, 'Emilia Romagna', 'Q')
session.load()
```

```python
lec = session.laps.pick_driver('LEC').pick_fastest()
ham = session.laps.pick_driver('HAM').pick_fastest()

delta_time, ref_tel, compare_tel = utils.delta_time(ham, lec)
# ham is reference, lec is compared

fig, ax = plt.subplots()
# use telemetry returned by .delta_time for best accuracy,
# this ensure the same applied interpolation and resampling
ax.plot(ref_tel['Distance'], ref_tel['Speed'],
        color=plotting.team_color(ham['Team']))
ax.plot(compare_tel['Distance'], compare_tel['Speed'],
        color=plotting.team_color(lec['Team']))

twin = ax.twinx()
twin.plot(ref_tel['Distance'], delta_time, '--', color='white')
twin.set_ylabel("<-- Lec ahead | Ham ahead -->")
plt.show()
```
